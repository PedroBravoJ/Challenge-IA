
# Avaliação de Fornecedores UrBuy



```python
# Importação das bibliotecas utilizadas
import pandas as pd
import numpy as np

# Biblioteca para normalização dos dados
from sklearn.preprocessing import MinMaxScaler

# Bibliotecas dos modelos de classificação utilizados
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier


# Biblioteca de treino do modelo
from sklearn.model_selection import train_test_split

# Biblioteca de visualização de imagens
from IPython.display import Image, display

# Biblioteca de números aleatórios utilizada para a
# criação da coluna Delivery Time
import random
```

```python
# Importando o dataset
df = pd.read_csv("/content/All Electronics.csv")
```

### Tratamento dos dados


```python
df.head()
```

```python
# Excluindo colunas desnecessárias

df = df.drop(columns=['sub_category', 'main_category', 'link', 'no_of_ratings', 'discount_price'])
df.head()
```

```python
# Excluindo linhas com valores nulos

df = df.dropna(subset=['name', 'ratings', 'actual_price'])
```

```python
# Tratamento da coluna "actual_price" para trabalhá-la como float

df['actual_price'] = df['actual_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace(',', '')
df['actual_price'] = df['actual_price'].astype(float)
df
```

```python
# Conversão de moeda Rupia Indiana (₹) para Real Brasileiro (R$)

df['actual_price'] = (df['actual_price'] * 0.06).round(2)
```

```python
# Criação da coluna "Delivery Time"

df['Delivery Time'] = [random.randint(1, 10) for _ in range(len(df))]
df
```

### Pesquisa por um produto específico

```python
# Função para busca de produto pelo nome

def search(user_search):
  filter = df[df["name"].str.contains(user_search, case=False)]

  if filter.empty:
    return "[ERRO] Produto não encontrado"
  else:
    return filter
```

```python
# Buscando produto e salvando em um novo dataset somente com as linhas encontradas

user_search = input("Digite o produto desejado: ")

df_produto = search(user_search)
df_produto
```

### Definição das métricas de avaliação e aplicação dos modelos de classificação


```python
# Normalização dos dados

scaler = MinMaxScaler()
df_produto[['actual_price', 'ratings', 'Delivery Time']] = scaler.fit_transform(df_produto[['actual_price', 'ratings', 'Delivery Time']])
```

```python
# Criação da coluna "Pontuação" com a média dos valores das outras 3 colunas, com diferentes pesos

weights = [0.4, 0.3, 0.3]  # Pesos para 'ratings', 'actual_price' e 'Delivery Time', respectivamente
df_produto['Pontuação'] = (weights[0] * df_produto['ratings'] +
                           weights[1] * df_produto['actual_price'] +
                           weights[2] * df_produto['Delivery Time'])
df_produto.head()
```

```python
# Separação das colunas em X e y e criação do modelo de classificação

X = df_produto[['actual_price', 'ratings', 'Delivery Time']]
y = df_produto['name']

# Dividir os dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Criação e treinamento do modelo KNeighborsClassifier

knn_model = KNeighborsClassifier(n_neighbors=5)
knn_model.fit(X_train, y_train)
```

```python
# Criação e treinamento do modelo RandomForestClassifier
forest_model = RandomForestClassifier(n_estimators=100, random_state=42)
forest_model.fit(X_train, y_train)
```

```python
# Input do usuário com os valores desejados para cada parâmetro e aplicação dos mesmos processos realizados com os dados anteriormente (normalização)

ratings = float(input("Digite a avaliação desejada (ratings) de 1 a 5: "))
actual_price = float(input("Digite o preço desejado (actual_price): "))
delivery_time = float(input("Digite o tempo de entrega desejado (Delivery Time): "))

novo_exemplo = pd.DataFrame({
    'ratings': [ratings],
    'actual_price': [actual_price],
    'Delivery Time': [delivery_time]
})

# Normalizar o novo exemplo
novo_exemplo[['actual_price', 'ratings', 'Delivery Time']] = scaler.transform(novo_exemplo[['actual_price', 'ratings', 'Delivery Time']])
```

```python
def calcular_distancias(df, exemplo):
    """Calcula a distância euclidiana entre cada linha do DataFrame e o exemplo."""
    return np.sqrt(((df - exemplo) ** 2).sum(axis=1))
```

```python
def mostrar_previsoes_knn(modelo, nome_modelo):
    # Encontrar os 5 vizinhos mais próximos do novo exemplo
    distances, indices = modelo.kneighbors(novo_exemplo[['actual_price', 'ratings', 'Delivery Time']])
    df_top5 = df_produto.iloc[indices[0]].sort_values(by='Pontuação', ascending=False).head(5)

    # Exibição dos resultados
    print(f"\nTop 5 recomendações - {nome_modelo}:")
    for index, row in df_top5.iterrows():
        print(f"Produto: {row['name']}")
        print(f"Preço: R${df.iloc[index]['actual_price']:.2f}")
        print(f"Tempo de Entrega: {row['Delivery Time']:.0f} dias")
        display(Image(url=row['image']))
        print("------------------------------------")
```

```python
def mostrar_previsoes_outros(modelo, nome_modelo):
    # Fazer a previsão para todos os produtos e calcular distâncias
    predicoes = modelo.predict(X)
    df_produto.loc[:, 'predicao'] = predicoes
    df_produto.loc[:, 'distancia'] = calcular_distancias(X, novo_exemplo[['actual_price', 'ratings', 'Delivery Time']].iloc[0])

    # Separando os resultados do df_produto
    df_top5 = df_produto.sort_values(by='distancia').head(5)

    # Exibição dos resultados
    print(f"\nTop 5 recomendações - {nome_modelo}:")
    for index, row in df_top5.iterrows():
        print(f"Produto: {row['name']}")
        print(f"Preço: R${df.iloc[index]['actual_price']:.2f}")
        print(f"Tempo de Entrega: {row['Delivery Time']:.0f} dias")
        display(Image(url=row['image']))
        print("------------------------------------")
```

### Apresentação dos resultados



```python
mostrar_previsoes_knn(knn_model, "KNN")
```

```python
mostrar_previsoes_outros(forest_model, "Random Forest")
```

# Autocrítica sobre o Desenvolvimento do Projeto de Avaliação de Produtos da Amazon

Durante o desenvolvimento deste projeto, cujo objetivo era avaliar e recomendar produtos de um dataset da Amazon de acordo com as preferências dos usuários, encontrei algumas dificuldades e desafios que impactaram o resultado final. A seguir, destaco os principais pontos de dificuldade enfrentados e os pontos positivos de aprendizado:

## Dificuldades Encontradas

1. **Avaliação em um Modelo de Redes Neurais**

   Um dos principais desafios foi a tentativa de aplicar um modelo de redes neurais. Embora as redes neurais possam oferecer grande precisão e flexibilidade na classificação e recomendação de produtos, não consegui implementá-las neste projeto. A complexidade adicional na configuração, treinamento e ajuste fino de redes neurais, combinada com a necessidade de um entendimento mais profundo dos parâmetros e arquitetura da rede, excedeu o tempo disponível para o desenvolvimento do projeto. Como resultado, a avaliação ficou limitada a modelos mais tradicionais, como K-Nearest Neighbors (KNN) e Random Forest.

2. **Integração com o Banco de Dados e Retorno ao Aplicativo**

   Outro desafio significativo foi a falta de integração do modelo desenvolvido com o banco de dados do projeto e a posterior devolução dos resultados ao aplicativo. Idealmente, o modelo treinado deveria ser capaz de acessar diretamente os dados em um banco de dados, processá-los e retornar as recomendações ao aplicativo em tempo real. No entanto, devido a limitações de tempo, essa integração crucial não foi realizada. A ausência dessa funcionalidade reduz a utilidade prática do modelo, já que a recomendação dos produtos não pode ser utilizada diretamente pelos usuários finais do aplicativo.

3. **Problemas com a Coluna de Nomes no Dataset**

   A qualidade e especificidade dos nomes dos produtos no dataset também apresentaram um problema significativo. Por exemplo, ao buscar pelo produto "mouse", resultados irrelevantes como "mouse pad" eram frequentemente incluídos. Isso se deve à falta de uma coluna de nome mais objetiva e específica no dataset, dificultando a filtragem precisa dos produtos desejados. Esse problema levou a resultados menos precisos e aumentou a necessidade de processamento adicional para limpar e filtrar os dados de maneira adequada.

4. **Média entre os Modelos KNN e Random Forest**

   Tentei aplicar uma média entre os dois modelos, KNN e Random Forest, na esperança de combinar os pontos fortes de cada um e melhorar a precisão das recomendações. No entanto, essa abordagem não produziu os resultados esperados. A média das previsões dos dois modelos não resultou em uma melhoria significativa na qualidade das recomendações e, em alguns casos, até piorou a precisão. Isso destaca a necessidade de um método mais robusto de ensemble ou combinação de modelos, que não foi possível implementar adequadamente dentro do prazo.

1. **Aplicação do Modelo KNN**

   Achei muito interessante e simples de aplicar o modelo KNN. Gostei bastante do resultado obtido, apesar de considerar que o Random Forest se apresentou mais preciso e condizente nos seus resultados. O KNN mostrou-se eficaz em diversas situações e proporcionou uma boa base para comparação com outros modelos mais complexos.

2. **Relevância do Tema do Projeto**

Por ser voltado a compras e vendas, mostrou-se bastante condizente com o mercado de trabalho atual, nos fazendo adquirir prática em processos muito utilizados em e-commerce, uma área bastante relevante e em crescimento no setor de tecnologia. Trabalhar com um dataset real da Amazon e desenvolver um sistema de recomendação proporcionou uma experiência prática valiosa.

3. **Comparação entre Modelos**

   A comparação entre diferentes modelos de classificação, como KNN e Random Forest, foi um exercício extremamente útil. Isso me permitiu entender melhor as vantagens e limitações de cada abordagem e como escolher o modelo mais adequado para diferentes tipos de problemas. Apesar das dificuldades, essa experiência contribuiu significativamente para meu desenvolvimento como profissional de ciência de dados.
